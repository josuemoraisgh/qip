```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import KFold, train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# Caminho para o arquivo xlsx local
file_path = './banco_dados.xlsx'
# Leitura do arquivo xlsx usando pandas
data = pd.read_excel(file_path, sheet_name='TDados')
```

```python
# Função para verificar problemas nos dados
def verificar_dados(X, y):
    print("Colunas constantes em X:", [c for c in X.columns if X[c].nunique() <= 1])
    print("Tem NaN em X:", np.isnan(X).any())
    print("Tem Inf em X:", np.isinf(X).any())
    print("Tem NaN em y:", np.isnan(y).any())
    print("Tem Inf em y:", np.isinf(y).any())

# Função para criar gráficos de aprendizado
def plotar_historico(historico, fold=None):
    plt.figure()
    plt.plot(historico.history['loss'], label='Treino')
    plt.plot(historico.history['val_loss'], label='Validação')
    plt.title(f'Evolução da Perda - Fold {fold}' if fold is not None else 'Evolução da Perda')
    plt.xlabel('Época')
    plt.ylabel('Perda')
    plt.legend()
    plt.show()

def criar_modelo(config):
    model = tf.keras.Sequential()

    # Primeira camada
    model.add(tf.keras.layers.Dense(config['units1'], activation=config['activation'],
                                     kernel_regularizer=tf.keras.regularizers.l2(config['l2_reg'])))

    # Opcional: BatchNormalization
    if config['use_batchnorm']:
        model.add(tf.keras.layers.BatchNormalization())

    # Opcional: Dropout ou AlphaDropout dependendo da ativação
    if config['activation'].lower() == 'selu':
        model.add(tf.keras.layers.AlphaDropout(config['dropout_rate']))
    else:
        model.add(tf.keras.layers.Dropout(config['dropout_rate']))

    # Camadas ocultas adicionais
    for units in [config['units2'], config['units3'], config['units4']]:
        model.add(tf.keras.layers.Dense(units, activation=config['activation'],
                                         kernel_regularizer=tf.keras.regularizers.l2(config['l2_reg'])))

        if config['use_batchnorm']:
            model.add(tf.keras.layers.BatchNormalization())

        if config['activation'].lower() == 'selu':
            model.add(tf.keras.layers.AlphaDropout(config['dropout_rate']))
        else:
            model.add(tf.keras.layers.Dropout(config['dropout_rate']))

    # Camada de saída
    model.add(tf.keras.layers.Dense(1, activation='linear'))

    # Compilar o modelo
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=config['learning_rate']),
                  loss='mse',
                  metrics=['mae'])

    return model

def treinar_kfold(X, y, config, n_splits=10):
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)
    fold_results = []

    for fold, (train_idx, val_idx) in enumerate(kf.split(X)):
        print(f"\nFold {fold + 1}/{n_splits}")

        # Divisão dos dados
        X_train, X_val = X.iloc[train_idx, :], X.iloc[val_idx, :]
        y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]

        # Normalização
        scaler = StandardScaler()
        X_train = scaler.fit_transform(X_train)
        X_val = scaler.transform(X_val)

        # Verificar variação em y
        if y_train.nunique() <= 1 or y_val.nunique() <= 1:
            print("Sem variação em y neste fold. Pulando...")
            continue

        # Criar o modelo
        model = criar_modelo(config)

        # Treinar o modelo
        history = model.fit(X_train, y_train,
                            validation_data=(X_val, y_val),
                            batch_size=max(8, len(X_train) // 10),  # batch_size adaptativo
                            epochs=50,
                            verbose=1)

        # Plotar histórico do treinamento
        plotar_historico(history, fold + 1)

        # Avaliar o modelo
        val_loss = model.evaluate(X_val, y_val, verbose=0)
        print(f"Val Loss: {val_loss[0]}, Val MAE: {val_loss[1]}")
        fold_results.append(val_loss)

    return fold_results

def treinar_e_avaliar(X, y, config, test_size=0.2):
    # Separar dados em treino e teste
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)

    # Separar uma porção de validação do conjunto de treino
    X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

    # Normalizar os dados
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_val = scaler.transform(X_val)
    X_test = scaler.transform(X_test)

    # Criar o modelo
    model = criar_modelo(config)
    early_stop = EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True)
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-7, verbose=1)

    # Treinar o modelo com validação
    history = model.fit(X_train, y_train,
                        validation_data=(X_val, y_val),
                        epochs=300,
                        batch_size=16,
                        callbacks=[early_stop, reduce_lr],
                        verbose=1)

    # Plotar histórico do treinamento
    plotar_historico(history)

    # Avaliar no conjunto de teste
    y_pred = model.predict(X_test).flatten()
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    print("\n[RESULTADOS NO TESTE]")
    print(f"MSE: {mse:.4f}")
    print(f"MAE: {mae:.4f}")
    print(f"R²: {r2:.4f}")

    return model


# Configurações ajustadas
# config = {
#     'units1': 32,
#     'units2': 16,
#     'units3': 8,
#     'units4': 4,
#     'activation': 'relu',  # ou 'selu' (nesse caso, AlphaDropout será usado)
#     'learning_rate': 1e-4,
#     'dropout_rate': 0.2,
#     'l2_reg': 1e-4,
#     'use_batchnorm': False
# }

config = {
    'units1': 64,
    'units2': 32,
    'units3': 16,
    'units4': 8,
    'activation': 'swish',
    'learning_rate': 1e-3,
    'dropout_rate': 0.1,
    'l2_reg': 1e-5,
    'use_batchnorm': True
}


# Pré-processamento inicial
X = data.drop(columns=['Alvo'], errors='ignore')
X = X.loc[:, X.nunique() > 1]  # Remover colunas constantes
y = data['Alvo'] if 'Alvo' in data else pd.Series(np.random.rand(len(data)))
y = y.apply(lambda x: 1 if x == 0.25 else 0) # testa apenas a ansiedade

# Verificar problemas nos dados
verificar_dados(X, y)

# Treinar com K-Fold
resultados_kfold = treinar_kfold(X, y, config, 10)
print("\nResultados finais por fold:", resultados_kfold)

# Treinar e avaliar no conjunto de teste
modelo_final = treinar_e_avaliar(X, y, config)
```
